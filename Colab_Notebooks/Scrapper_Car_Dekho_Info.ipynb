```python
# importing builtin libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.chrome.options import Options

# importing my libraries
import import_ipynb
import Scrapper_Car_Dekho_Links as scrapper
```

```python
# this method fetches the car name and splits the information according to the requirements

def get_car_name(driver):
    car_name = driver.find_element(By.CLASS_NAME, 'vehicleName')
    car_name = car_name.text.split()
    car_year, car_brand, *car_model = car_name
    car_year = int(car_year)
    car_model = ' '.join(car_model)

    return [car_brand, car_model, car_year]
```

```python
# fetching car price

def get_car_price(driver):
    car_price = driver.find_element(By.CLASS_NAME, 'vehiclePrice')
    price = car_price.text.split(' ')
    price = float(price[1])

    return price
```

```python
#fetching overview and specification of the car

def car_details(driver):
    ul = driver.find_elements(By.XPATH, "//ul[@class='gsc_row detailsList']")
    [overview, specifications] = ul
    overview = overview.text.split('\n')
    specifications = specifications.text.split('\n')

    details = overview + specifications

    return details
```

```python
# getting the list of car details without passing any additional arguments

def get_car_details():
    return car_details(driver)
```

```python
# checking if the key exists of the particular car attribute
def key_exists(key):
    details = get_car_details()
    for i in range(0,len(details),2):
        if details[i] == key:
            return True
    return False
```
